# 📝 Сборщик фраз Петровича
Добавляй фразы в список ниже и запускай ячейку — они сохранятся в `phrases.txt`.

```python
PHRASES_FILE = "phrases.txt"

def load_phrases():
    import os
    if not os.path.exists(PHRASES_FILE):
        return []
    with open(PHRASES_FILE, "r", encoding="utf-8") as f:
        return [line.strip() for line in f if line.strip()]

def save_phrases(new_phrases):
    existing = set(load_phrases())
    added = 0
    with open(PHRASES_FILE, "a", encoding="utf-8") as f:
        for phrase in new_phrases:
            phrase = phrase.strip()
            if phrase and phrase not in existing:
                f.write(phrase + "\n")
                existing.add(phrase)
                added += 1
    return added

existing = load_phrases()
print(f"Уже в базе: {len(existing)} фраз")
```
